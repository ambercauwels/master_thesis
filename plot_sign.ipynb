Plotting functions to make nice significance and limit plots

```python
import numpy as np
import matplotlib.pyplot as plt
import mplhep as hep


def make_layout(var):
    xlabels = {"eta" : "$\eta$", "pt": "$p_T$ (GeV)", "mass": "mass (GeV)", 'M_sd': r'$M_{SD}$',
                "phy": "$\phi$", "CA": r"$\Delta^{t}_{\eta}$", "dR": r"$\Delta{r}$", "t32": r"$\tau_{32}$", 'n': 'N', 'Ht': r'H$_T$', 's': r'expected significance $\sigma$'}
    
    fig, ax = plt.subplots(figsize=(7, 5))
    hep.cms.label(ax=ax, fontsize=16, llabel='Simulation Internal', rlabel=r'138 fb$^{-1}$ (13 TeV)')
    ax.set_xlabel(xlabels[var], fontsize=16)
    ax.set_xlim(0, 10)
    #ax.set_xscale('log')
    return ax
```

```python
p = [r'$t\bar{t}Z \rightarrow$ 2l', r'$t\bar{t}Z\rightarrow$ 3l', r'$t\bar{t}W \rightarrow$ 2l']
y_pos = np.arange(len(p))
sig = {'ttZ2l': 3.53244, 'ttZ3l': 8.87954, 'ttW': 4.8352}

ax = make_layout('s')
ax.barh(y_pos, sig.values(), align='center', color=["#3f90da", "#3f90da", "#bd1f01"], alpha=0.8)
ax.set_yticks(y_pos)
ax.set_yticklabels(p, fontsize=16)
ax.invert_yaxis()  
for i, v in enumerate(sig.values()):
    ax.text(v+0.2, i , f'{v:.2f}', fontsize=13)

plt.show()
```

```python
p = [r'$(m_A, m_H)$=(2100, 2000) GeV', r'$(m_A, m_H)$=(2000, 1500) GeV', r'$(m_A, m_H)$=(1000, 850) GeV']
y_pos = np.arange(len(p))

r = [550.2370, 6.4772, 5.9405]
r_p95 = [1297.2234, 14.2698, 11.9534]
r_m95 = [253.0547, 3.1586, 3.0847]
r_p68 = [861.9323, 9.7950, 8.5882]
r_m68 = [359.4100, 4.3740, 4.1768]


fig, ax = plt.subplots(figsize=(7, 5))
hep.cms.label(ax=ax, fontsize=16, llabel='Simulation Internal', rlabel=r'138 fb$^{-1}$ (13 TeV)')
ax.vlines(r, ymin=y_pos, ymax=y_pos+1, color='black',  ls='dashed', label='expected')
ax.vlines(1, ymin=0, ymax=3, color='r')

#ax.fill_betweenx(y=np.linspace(0, 3, 4), x1=r_m95, x2=r_p95, color='gold', interpolate=True, step='post', label=r'95% expected')

ax.fill_betweenx([0, 1], x1=r_m95[0], x2=r_p95[0], color='gold', interpolate=True, step='post', label=r'95% expected')
ax.fill_betweenx([1, 2], x1=r_m95[1], x2=r_p95[1], color='gold', interpolate=True, step='post')
ax.fill_betweenx([2, 3], x1=r_m95[2], x2=r_p95[2], color='gold', interpolate=True, step='post')



ax.fill_betweenx([0, 1], x1=r_m68[0], x2=r_p68[0], color='limegreen', interpolate=True, step='post', label=r'68% expected')
ax.fill_betweenx([1, 2], x1=r_m68[1], x2=r_p68[1], color='limegreen', interpolate=True, step='post')
ax.fill_betweenx([2, 3], x1=r_m68[2], x2=r_p68[2], color='limegreen', interpolate=True, step='post')


ax.set_yticks(y_pos+0.5)
ax.set_yticklabels(p, fontsize=16)
ax.set_ylim(0, 3)
ax.set_xscale('log')
ax.set_xlabel(r'95% CL limit on r=$\sigma_{obs}$/$\sigma_{theory}$', fontsize=16)
ax.legend(fontsize=14, loc='upper left')

ax.invert_yaxis()  
```

```python
p = [r'$t\bar{t}Z \rightarrow$ 2l', r'$t\bar{t}Z\rightarrow$ 3l', r'$t\bar{t}W \rightarrow$ 2l']

y_pos = np.arange(len(p))

r = [1, 1, 1]
r_p95 = [1.29, 1.18, 1.24]
r_m95 = [0.72, 0.84, 0.77]



fig, ax = plt.subplots(figsize=(7, 5))
hep.cms.label(ax=ax, fontsize=16, llabel='Simulation Internal', rlabel=r'138 fb$^{-1}$ (13 TeV)')
ax.vlines(r, ymin=y_pos, ymax=y_pos+1, color='black',  ls='dashed', label='expected')

#ax.fill_betweenx(y=np.linspace(0, 3, 4), x1=r_m95, x2=r_p95, color='gold', interpolate=True, step='post', label=r'95% expected')

ax.fill_betweenx([0.1, 0.9], x1=r_m95[0], x2=r_p95[0], color='skyblue', interpolate=True, step='post', label=r'95% CL. expected')
ax.fill_betweenx([1.1, 1.9], x1=r_m95[1], x2=r_p95[1], color='skyblue', interpolate=True, step='post')
ax.fill_betweenx([2.1, 2.9], x1=r_m95[2], x2=r_p95[2], color='skyblue', interpolate=True, step='post')



ax.set_yticks(y_pos+0.5)
ax.set_yticklabels(p, fontsize=16)
ax.set_ylim(0, 3)
ax.set_xlabel(r'$\hat{r}=\sigma_{obs}/\sigma_{theory}$', fontsize=16)
ax.legend(fontsize=14, loc='upper left')

ax.invert_yaxis()  
```
